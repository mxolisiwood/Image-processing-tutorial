Resizing the image using Pillow library

```python
from PIL import Image
```

```python
img = Image.open('/content/dog.jpg')
img_resize = img.resize((200,200))
```

```python
img_resize.save('dog_image_resized.jpg')
```

```python
img_res = mpimg.imread('/content/dog_image_resized.jpg')
```

```python
img_res_plot = plt.imshow(img_res)
plt.show() 
```

```python
print(img_res.shape)
```

```python
print(img_res)
```

Converting RGB image to Grayscale image using OpenCV

```python
import cv2
```

```python
img = cv2.imread('/content/dog.jpg')
```

```python
type(img)
```

```python
print(img.shape)
```

```python
grayscale_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
print(grayscale_image)
```

```python
type(grayscale_image)
```

```python
print(grayscale_image.shape)
```

```python
#cv2.imshow() will only work on other IDEs but not google colab

from google.colab.patches import cv2_imshow
```

```python
cv2_imshow(grayscale_image)
```

```python
#saving the gray image
cv2.imwrite('dog_grayimage.jpg', grayscale_image)
```
